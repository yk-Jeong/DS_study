### 5.3. 데이터 전처리 - tidy data 만들기 

- melt: 행에 있는 데이터를 열로 옮김

```python
# 옵션 살펴보기 

df_raw.melt?
```

```python
# id_vars: id로 사용할 열 지정, value_vars: value로 들어갈 열 지정 

df_raw.melt(id_vars = ['국가(대륙)별', '상품군별', '판매유형별'])
```

```python
df = df_raw.melt(id_vars = ['국가(대륙)별', '상품군별', '판매유형별'], var_name = '기간', value_name = '백만원')
df.shape
```

```python
df.info()
```

### 5.4 데이터 전처리 

```python
df['기간']
```

```python
"2019.4/4".split()
```

```python
"2019.4/4".split('.')[0]
```

```python
# map 함수 사용
# df['기간'].map?

df['기간'].map(lambda x: int(x.split('.')[0]))
```

```python
# 기간에서 연도 분리

df['연도'] = df['기간'].map(lambda x: int(x.split('.')[0]))
df.head()
```

```python
# 기간에서 분기 분리

df['분기'] = df['기간'].map(lambda x: int(x.split('.')[1].split('/')[0]))
df.head()
```

```python
# 결측치 수정

df['백만원'].replace('-', pd.np.nan)
```

```python
# 형 변경

df['백만원'] = df['백만원'].replace('-', pd.np.nan).astype(float)
df['백만원']
```

```python
df.head()
```

```python
# 불필요한 데이터 삭제 

df = df[(df['국가(대륙)별'] != '합계') & (df['상품군별'] != '합계')].copy()
df.info()
```

```python
# 결측치 확인 

df.isnull().sum()
```

### 5.5 증가추세 시각화

- sns.lineplot() / sns.relplot()
- plt.legend()

```python
df_total = df[df['판매유형별'] == '계'].copy()
df_total.head()
```

```python
sns.lineplot(data = df_total, x='연도', y='백만원')
```

```python
sns.relplot(data = df_total, x = '연도', y='백만원', hue = '상품군별', kind='line')
```

```python
sns.relplot(data = df_total, x = '연도', y='백만원', hue = '상품군별', col='상품군별', kind = 'line', col_wrap = 4)
```
